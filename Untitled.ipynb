***Importar a base de dados***

```python
import pandas as pd 
```

```python
tabela_vendas = pd.read_excel ('Vendas.python.xlsx')
```

***- Visualizar a base de dados***

```python
tabela_vendas
```

```python
tabela_vendas [['ID Loja','Valor Final']]
```

***- Faturamento por loja***

```python
faturamentos_final_lojas = tabela_vendas[['ID Loja', 'Valor Final']].groupby('ID Loja').sum()
```

```python
faturamentos_final_lojas
```

***- Quantidade de produtos vendidos por loja***

```python
quantidades_produtos_lojas = tabela_vendas[['ID Loja', 'Quantidade']].groupby('ID Loja').sum()
```

```python
quantidades_produtos_lojas
```

***- Ticket médio por produtos em cada loja***

```python
ticket_medio_produtos_lojas = (faturamentos_final_lojas['Valor Final'] / quantidades_produtos_lojas['Quantidade']).to_frame()
ticket_medio_produtos_lojas
ticket_medio_produtos_lojas = ticket_medio_produtos_lojas.rename(columns={0: 'Ticket Médio'})
ticket_medio_produtos_lojas
```

***- Envio de Email com relatório***

```python
import win32com.client as win32
outlook = win32.Dispatch('outlook.application')
mail = outlook.CreateItem(0)
mail.To = 'nobrugonferr@gmail.com'
mail.Subject = 'Projeto Python Relatorio de Vendas '
mail.HTMLBody = f'''
<p>Bom dia, Prezados</p>

<p>Segue o relatório de vendas por cada loja.</p>

<p>Faturamentos:</p>
{faturamentos_final_lojas.to_html(formatters={'Valor Final': '£{:,.2f}'.format})}


<p>Quantidade Vendida:</p>
{quantidades_produtos_lojas.to_html()}


<p>Ticket Medio dos Produtos por loja:</p>
{ticket_medio_produtos_lojas.to_html(formatters={'Ticket Médio': '£{:,.2f}'.format})}


<p>Qualquer duvida estou a disposição.</p>

<p>Att, Bruno Ferreira</p>
'''

mail.Send()
```
